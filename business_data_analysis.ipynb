```python
# ================================
# Business Data Analysis Project
# ================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display settings
pd.set_option('display.max_columns', None)
sns.set(style="whitegrid")

# ----------------
# Load Dataset
# ----------------
# Sample dataset (you can replace with real data later)
data = {
    "Order_ID": range(1, 21),
    "Region": ["North", "South", "East", "West"] * 5,
    "Product": ["A", "B", "C", "D", "A"] * 4,
    "Sales": [200, 300, 150, 400, 250, 220, 330, 180, 410, 270,
              210, 320, 160, 390, 260, 230, 340, 170, 420, 280],
    "Month": ["Jan", "Feb", "Mar", "Apr", "May"] * 4
}

df = pd.DataFrame(data)

# ----------------
# Data Inspection
# ----------------
df.head()
```

```python
# Check basic info
df.info()

# Check for missing values
df.isnull().sum()
```

```python
# ----------------
# Data Cleaning
# ----------------
# (No missing values here, but structure shown for real-world use)

df["Sales"] = df["Sales"].astype(float)

# ----------------
# Exploratory Data Analysis
# ----------------

# Total sales by region
sales_by_region = df.groupby("Region")["Sales"].sum().reset_index()

# Total sales by product
sales_by_product = df.groupby("Product")["Sales"].sum().reset_index()

sales_by_region, sales_by_product
```

```python
# ----------------
# Visualisation
# ----------------

plt.figure(figsize=(8,5))
sns.barplot(data=sales_by_region, x="Region", y="Sales")
plt.title("Total Sales by Region")
plt.show()

plt.figure(figsize=(8,5))
sns.barplot(data=sales_by_product, x="Product", y="Sales")
plt.title("Total Sales by Product")
plt.show()
```

```python
# ----------------
# Business Insights
# ----------------

print("Key Business Insights:")
print("- The highest revenue comes from the region with the tallest bar.")
print("- Product performance varies significantly across categories.")
print("- These insights can guide inventory and sales strategy.")
```
